```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras import models
```

```python
model = models.load_model("model from kaggle.h5")
```

```python
def image_predict(path):
    #loading image from given path and changing it in array
    test_image = image.load_img(path,target_size=(224,224,3))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image,axis=0)
    print(test_image)
    #making prediction
    result = model.predict(test_image)
    print(result)
    #now checking result
    result = np.argmax(result)
    print(result)
    if result == 0:
        prediction = 'Bacterial leaf blight'
    elif result == 1:
        prediction = 'Brown spot'
    else:
        prediction = 'Leaf smut'
    return prediction
```

```python
print("Correct Bacterial-leaf-blight And Predict :",image_predict("datasets/rice_Bacterial-leaf-blight.jpg"))
print("Correct Leaf_smut And Predict :",image_predict("datasets/rice_Leaf_smut.jpg"))
print("Correct Leaf_smut And Predict :",image_predict("datasets/rice_leaf_smut1.jpg"))
print("Correct brown_spot And Predict :",image_predict("datasets/rice_brown_leaf.jpg"))
```
